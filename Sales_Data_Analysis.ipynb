```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv(('shopping_trends_updated.csv'))
sales_data.head()
```

```python
sales_data.groupby('Payment Method')['Payment Method'].count().sort_values(ascending=False)
```

```python
base_color = sb.color_palette()[0]
method = sales_data['Payment Method'].value_counts().index
sb.countplot(data=sales_data, y="Payment Method", color = base_color, order = method)
```

```python
sales_data.groupby("Item Purchased")['Item Purchased'].count().sort_values(ascending=False)
```

```python
base_color = sb.color_palette()[3]
method = sales_data['Item Purchased'].value_counts().index
sb.countplot(data=sales_data, y="Item Purchased", color = base_color, order = method)
```

```python
sales_data.Gender.value_counts()
```

```python
sales_data.Gender.unique()
```

```python
sales_data[(sales_data['Gender'] == 'Female')].groupby('Item Purchased')['Item Purchased'].count().sort_values(ascending=False)
```

```python
base_color = sb.color_palette()[1]

method = sales_data['Gender'].value_counts().index

sb.countplot(data = sales_data, y='Gender', color = base_color, order = method)
```

```python
location = sales_data['Location'].value_counts().head(10).index

sb.countplot(data = sales_data, y = 'Location', color = base_color, order = location)
```
